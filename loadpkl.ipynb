```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open('my_list.pkl', 'rb') as file:
    loaded_list = pickle.load(file)
```

```python
plt.figure()

idx =  7
plt.scatter([obj for obj in loaded_list['time']], [obj[idx] for obj in loaded_list['desired']], label='desired')
plt.scatter([obj for obj in loaded_list['time']], [obj[idx] for obj in loaded_list['state']], label='state')    

plt.title('Right Knee')
plt.xlabel('Time (seconds)')
plt.ylabel('Joint Position (radians)')
plt.legend(loc='upper right')
plt.savefig('/home/dhruv/Desktop/Digit/rk.png', dpi=300)
plt.show()
```
